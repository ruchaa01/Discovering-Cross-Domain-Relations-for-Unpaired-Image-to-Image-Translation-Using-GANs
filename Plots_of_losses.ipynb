```
import csv
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```
dt = [('epoch #', 'S64'),
      ('GEN Loss A', 'float'),
      ('GEN Loss B', 'float'),
      ('FM Loss A', 'float'),
      ('FM Loss B', 'float'),
      ('RECON Loss A', 'float'),
      ('RECON Loss B', 'float'),
      ('DIS Loss A', 'float'),
      ('DIS Loss B', 'float')]
```

```
d = np.loadtxt('loss_data.csv',delimiter=',',dtype=dt,skiprows=1) # skip the header row
```

```
print(d['GEN Loss A'][:5])  #Test
```

```
epochs = np.asarray(range(1000))
```

```
print(epochs.shape)
print(d['GEN Loss A'].shape)
```
